## P6 ('xml')

```python
# Para leer 'xml' primero lo abrimos con ET.parse
tree = ET.parse('Ped06.xml')
root = tree.getroot()

# Obtenemos una lista de los nombres de las columnas haciendo un bucle
column_names = []
for child in root.findall('.//record/*'):
    if child.tag not in column_names:
        column_names.append(child.tag)

# Creamos una lista de diccionarios con los datos de cada fila
data = []
for record in root.findall('.//record'):
    row = {}
    for child in record:
        row[child.tag] = child.text
    data.append(row)

# Creamos el DataFrame con la lista de diccionarios
df_p6 = pd.DataFrame(data, columns=column_names)
df_p6
```

# Filtro columnas

```python
# Seleccionamos las columnas que nos interesa y las metemos en una lista
lista_columnas = ['EjercicioPedido','SeriePedido','NumeroPedido','Orden','FechaRegistro','FechaPedido','CodigoArticulo',
                  'CodigodelCliente','CodigoProveedor','FechaCaduca','FechaEntrega','FechaTope','Estado','%Iva','%Recargo',
                 'UnidadesPedidas','UnidadesServidas','Precio','PrecioCoste','%Descuento','ImporteCoste','ImporteBruto']
```

```python
# Usamos una función para que nos devuelva el dataframe sólo con las columnas seleccionadas
def filtrar_columnas(df, lista_columnas):
    """
    Esta función toma un DataFrame y una lista de columnas como entrada y devuelve un nuevo DataFrame
    que contiene solo las columnas especificadas en la lista.
    """
    columnas_para_eliminar = set(df.columns) - set(lista_columnas)
    df_filtrado = df.drop(columns=columnas_para_eliminar)
    return df_filtrado
```

# Comprobación de valores nulos

```python
# Usaremos una función para saber el porcentaje de valores nulos de cada columna
def porcentaje_nulos(df):
    """
    Esta función toma un DataFrame como entrada y devuelve un DataFrame que muestra el porcentaje de valores nulos
    en cada columna.
    """
    total_filas = len(df)
    nulos_por_columna = df.isna().sum()
    porcentaje_nulos_por_columna = round((nulos_por_columna / total_filas) * 100, 2)
    df_resultado = pd.DataFrame({'Columna': df.columns, '% Nulos': porcentaje_nulos_por_columna})
    return df_resultado

# Podemos comprobarlo con el dataframe unido o cada uno por separado en función de los que nos interese
```

```python
nulos_columna_unidos = porcentaje_nulos(df_unidos)
nulos_columna_unidos
```
